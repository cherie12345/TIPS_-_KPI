```python
from datetime import datetime, timedelta
from dotenv import load_dotenv
from keras.callbacks import EarlyStopping, History
from keras.layers import BatchNormalization, Concatenate, Dense, Dropout, Embedding, Flatten, Input
from keras.optimizers import Adam
from logging.handlers import RotatingFileHandler
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder
from tensorflow import keras
from tqdm import tqdm

import boto3
import dask.dataframe as dd
import io
import joblib
import logging
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pymysql
import random
import tensorflow as tf
import traceback


class ShopperKPI():
    def __init__(self, ctg, model, year, month, day):
        dotenv_path = Path("/home/ailee/ailee-data-app/.env")
        load_dotenv(dotenv_path=dotenv_path)
        AWS_ACCESS_KEY_ID = os.getenv("AWS_ACCESS_KEY_ID")
        AWS_SECRET_ACCESS_KEY_ID = os.getenv("AWS_SECRET_ACCESS_KEY_ID")
        self.DB_ID = os.getenv("DB_ID")
        self.DB_PASSWORD = os.getenv("DB_PASSWORD")
        self.s3 = boto3.client('s3', aws_access_key_id=AWS_ACCESS_KEY_ID, aws_secret_access_key=AWS_SECRET_ACCESS_KEY_ID)
        self.aws_key = {'key': AWS_ACCESS_KEY_ID, 'secret': AWS_SECRET_ACCESS_KEY_ID}
        self.log_path = Path('/home/ailee/ailee-data-app/log')
        self.ctg = ctg
        self.logger = self.create_logger()
        # if(ctg == 'pet'):
        #     # model_path = Path("/home/ailee/ailee-data-app/model/lightgbm_230925.pkl")
        #     model_path = Path("/home/ailee/ailee-data-app/model/hp_tuned_best_model.h5")
        # elif(ctg == 'car'):
        #     model_path = Path("/home/ailee/ailee-data-app/model/catboost_car_bestloss.pkl")
        # elif(ctg == 'living'):
        #     model_path = Path("/home/ailee/ailee-data-app/model/xgboost_model_6month_6input_living_1208.pkl")

        self.model = model

        self.now = datetime(year, month, day)
        self.year = self.now.year
        self.month = self.now.month
        self.week = self.now.strftime("%V")
        self.end_date = self.now-timedelta(days=1)
        self.start_date = self.now-timedelta(days=180)
        self.label_encoder = joblib.load('/home/ailee/ailee-v1-ai/DNN/label_encoder.joblib')


    def update_table(self, df, table_name='kpi_predict_6months'):
        conn = self.connect_database(db_name='ailee', user=self.DB_ID, password=self.DB_PASSWORD)

        with conn:
            with conn.cursor() as cursor:
                # upsert
                columns = ', '.join(df.columns)
                values = ', '.join(['%s'] * len(df.columns))
                query = f"INSERT INTO ailee.{table_name} ({columns}) VALUES ({values}) " \
                        f"ON DUPLICATE KEY UPDATE " \
                        f"{', '.join([f'{col} = VALUES({col})' for col in df.columns if col != 'primary_key_column'])}"

                data_tuples = [tuple(row) for row in df.values]
                cursor.executemany(query, data_tuples)

            conn.commit()

    def create_logger(self, ):
        if not os.path.exists(self.log_path):
            os.mkdir(self.log_path)

        logger = logging.getLogger(__name__)
        logger.setLevel(logging.DEBUG)

        if(len(logger.handlers) == 3):
            return logger
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

        handler1 = logging.StreamHandler()
        handler1.setLevel(logging.INFO)
        handler1.setFormatter(formatter)

        handler2 = RotatingFileHandler(f'{self.log_path}/shopperkpi_debug.log', maxBytes=100*1024*1024, backupCount=5)
        handler2.setLevel(logging.DEBUG)
        handler2.setFormatter(formatter)

        handler3 = RotatingFileHandler(f'{self.log_path}/shopperkpi_error.log', maxBytes=100*1024*1024, backupCount=5)
        handler3.setLevel(logging.ERROR)
        handler3.setFormatter(formatter)

        logger.addHandler(handler1)
        logger.addHandler(handler2)
        logger.addHandler(handler3)

        return logger

    
    def connect_database(self, db_name, user, password):
        conn = pymysql.connect(
            host='aileedb-v1.cluster-cra214iluhkt.ap-northeast-2.rds.amazonaws.com',
            user=user,
            password=password,
            db=db_name,
            charset='utf8mb4'
        )

        return conn


    def get_kpi(self, data):
        r2 = r2_score(data['purchaseCnt'], data['predict_sales'])
        rmse = np.sqrt(mean_squared_error(data['purchaseCnt'], data['predict_sales']))
        mae = mean_absolute_error(data['purchaseCnt'], data['predict_sales'])
        mape = mean_absolute_percentage_error(data['purchaseCnt'], data['predict_sales'])
        mpe = np.mean((data['purchaseCnt']-data['predict_sales'])/data['purchaseCnt'])

        return pd.Series(dict(r2 = r2, rmse = rmse, mae = mae, mape = mape, mpe = mpe))


    def get_data_from_s3(self, bucket_name, folder_path):
        combined_df = pd.DataFrame()

        paginator = self.s3.get_paginator('list_objects_v2')
        pages = paginator.paginate(Bucket=bucket_name, Prefix=folder_path)
        for page in pages:
            for obj in page['Contents']:
                try:
                    file_path = obj['Key']
                    r = self.s3.get_object(Bucket=bucket_name, Key=file_path)
                    df = pd.read_parquet(io.BytesIO(r['Body'].read()))
                    combined_df = pd.concat([combined_df, df])
                except:
                    pass

        return combined_df
    
    
    def get_target_list(self,):
        conn = self.connect_database(db_name='ailee', user=self.DB_ID, password=self.DB_PASSWORD)
        
        query1 = """
        SELECT
            id
        FROM
            ailee.kpi_target_item
        """

        query2 = """
        SELECT 
            id
        FROM
            ailee.review_smartstore_crawling_target;
        """

        with conn:
            with conn.cursor(pymysql.cursors.DictCursor) as cursor:
                cursor.execute(query1)
                result1 = pd.DataFrame(cursor.fetchall())
                cursor.execute(query2)
                result2 = pd.DataFrame(cursor.fetchall())
                
        result = pd.concat([result1, result2])

        return result
    

    def get_info(self,):
        combined_df = pd.DataFrame()
        bucket_name = 'ailee-crawling-data'
        prefix = f'new-naver/checkout/{self.ctg}/raw/'

        folder_paths = []

        subdirectories = self.list_subdirectories(bucket_name, prefix)

        self.logger.info(subdirectories)

        for subdirectory in subdirectories:
            subdirectories_depth2 = self.list_subdirectories(bucket_name, prefix=subdirectory)
            folder_paths.extend(subdirectories_depth2)

        #self.logger.info(folder_paths)

        for folder_path in folder_paths:
            self.logger.info(folder_path)
            info = dd.read_parquet(f's3://ailee-crawling-data/{folder_path}*.parquet',
                                   storage_options = self.aws_key)
            
            #self.logger.info(info.columns)
            if('crawling_date' not in info.columns):
                self.logger.info(info.columns)
                continue
            else:
                pass
            info = info[['id', 'originalMallProductId', 'rank', 'price', 'crawling_date',
                    'category1Id', 'category2Id', 'category3Id', 'category4Id',
                        'category1Name', 'category2Name', 'category3Name', 'category4Name',
                        'purchaseCnt']]
            
            info = info.compute()

            combined_df = pd.concat([combined_df, info])

        combined_df['crawling_date'] = pd.to_datetime(pd.to_datetime(combined_df['crawling_date']).dt.strftime("%Y-%m-%d"))
        combined_df = combined_df.drop_duplicates(subset=['id', 'crawling_date'])
    

        self.logger.info(combined_df.shape)
        self.logger.info(combined_df.describe())
        self.logger.info(combined_df.info())

        combined_df.columns = ['id', 'origin_product_no', 'rank', 'price', 'create_date',
            'category1_id', 'category2_id', 'category3_id', 'category4_id', 
            'category1_name', 'category2_name', 'category3_name', 'category4_name', 'purchaseCnt']
        
        combined_df['origin_product_no'] = combined_df['origin_product_no'].apply(str)#.replace('', 0).apply(int)
        combined_df['rank'] = combined_df['rank'].apply(int)
        # combined_df['price'] = combined_df['price'].apply(int)

        return combined_df
    
    
    def list_subdirectories(self, bucket_name, prefix):
        # list_objects_v2를 사용하여 폴더 내 객체의 리스트를 가져옴
        response = self.s3.list_objects_v2(Bucket=bucket_name, Prefix=prefix, Delimiter='/')

        # CommonPrefixes에서 하위 폴더 목록 추출
        subdirectories = [prefix['Prefix'] for prefix in response.get('CommonPrefixes', [])]

        return subdirectories


    def get_review(self, ):
        combined_df = pd.DataFrame()
        bucket_name = 'ailee-crawling-data'
        prefix = 'naver/review/smartstore/data/store/'

        folder_paths = []

        subdirectories = self.list_subdirectories(bucket_name, prefix)

        for subdirectory in subdirectories:
            subdirectories_depth2 = self.list_subdirectories(bucket_name, prefix=subdirectory)
            folder_paths.extend(subdirectories_depth2)


        for folder_path in folder_paths:
            try:
                review = dd.read_csv(f's3://ailee-crawling-data/{folder_path}*.csv',
                                storage_options = self.aws_key,
                                dtype={'modifyDate': 'object',
                                        'reviewCommentIds': 'object',
                                        'reviewComments': 'object',
                                        'eventTitle': 'object',
                                        'detailCategorizeCategoryId': 'float64',
                                        'reviewInspectionPolicyReason': 'object'})
                review = review[['id', 'productNo', 'productName', 'originProductNo', 'createDate']]
                review = review.compute()
                review = review.drop_duplicates(subset=['id'])
                review['createDate'] = pd.to_datetime(review['createDate']).dt.strftime("%Y-%m-%d")

                combined_df = pd.concat([combined_df, review])
            except:
                pass

        combined_df = combined_df.drop_duplicates(subset=['id'])
        combined_df = combined_df.groupby(['originProductNo', 'createDate']).count()[['id']].reset_index(drop=False)
        combined_df.columns = ['origin_product_no', 'create_date', 'daily_review_count']

        return combined_df
    

    def create_daterange(self, review):
        date_range = pd.date_range(
            start=datetime(self.start_date.year, self.start_date.month, self.start_date.day),
            end=datetime(self.end_date.year, self.end_date.month, self.end_date.day),
            freq='D'
        )

        origin_product_nos = review['origin_product_no'].unique()
        fillna_idx = (
            pd.MultiIndex
            .from_product(
                iterables=[origin_product_nos, date_range],
                names=['origin_product_no', 'create_date']
            )
        )

        review['create_date'] = pd.to_datetime(review['create_date'])

        min_review_date = review.groupby(['origin_product_no']).min().reset_index(drop=False)

        review = review.set_index(['origin_product_no', 'create_date'])

        review_count = review.reindex(fillna_idx, fill_value=0).reset_index(drop=False)

        full_review_ids = min_review_date[min_review_date.create_date <= self.start_date.strftime("%Y-%m-%d")]['origin_product_no'].unique()


        return review_count, full_review_ids
    

    def get_ailee_gmv(self, level):
        query = f"""
        WITH ctg AS 
        (	SELECT DISTINCT
                bcmi.ref_category1_id,
                bcmi.ref_category2_id,
                bcmi.ref_category3_id,
                bcmi.ref_category4_id,
                bcmi.category1_id,
                bcmi.category2_id,
                bcmi.category3_id,
                bcmi.category4_id
            FROM
                ailee.base_category_mapping_info AS bcmi
        )

        SELECT DISTINCT
            SUM(ami.gmv) AS gmv,
            ctg.ref_category{level}_id AS category{level}_id
        FROM
            ailee.analysis_map_info AS ami
        LEFT JOIN ctg
            ON ami.category4_id = ctg.category4_id
        WHERE
            ami.week = 202346
            AND
                ami.category1_id = 19
        GROUP BY ctg.ref_category{level}_id
        """

        conn = self.connect_database(db_name='ailee', user=self.DB_ID, password=self.DB_PASSWORD)

        with conn:
            with conn.cursor(pymysql.cursors.DictCursor) as cursor:
                cursor.execute(query)
                data = pd.DataFrame(cursor.fetchall())

        return data
    

    def get_actual_sales(self, info):
        actual_sales = info.sort_values(by=['create_date'], ascending=True)
        actual_sales.drop_duplicates(subset=['origin_product_no'], keep='last', inplace=True)
        actual_sales = info[['origin_product_no', 'purchaseCnt']]
        actual_sales.columns = ['origin_product_no', 'purchase_cnt']
        actual_sales['origin_product_no'] = actual_sales['origin_product_no'].replace('', 0).apply(int)
        actual_sales.drop_duplicates(subset=['origin_product_no'], keep='last', inplace=True)

        return actual_sales


    # def predict(self, df):
    #     input_features = ['daily_review_count',
    #                       'rank',
    #                       'price',
    #                     #   'category1_id',
    #                     #   'category2_id',
    #                     #   'category3_id',
    #                     #   'category4_id'
    #                       ]
    #     # df['category1_id'] = df['category1_id'].fillna('').replace({'':0}).apply(int)
    #     # df['category2_id'] = df['category2_id'].fillna('').replace({'':0}).apply(int)
    #     # df['category3_id'] = df['category3_id'].fillna('').replace({'':0}).apply(int)
    #     # df['category4_id'] = df['category4_id'].fillna('').replace({'':0}).apply(int)
        
    #     df['predict_sales'] = self.model.predict(df[input_features])
    #     df['predict_sales'] = np.round(df['predict_sales']/3)

    #     return df
    
    # def predict(self, df):
    #     enc = joblib.load('/home/ailee/ailee-v1-ai/DNN/label_encoder.joblib')
    #     df['origin_product_no'] = df['origin_product_no'].apply(str)
    #     df = df[df.origin_product_no.isin(enc.classes_)]

    #     df['id'] = enc.transform(df['origin_product_no'])
    #     input_features = ['daily_review_count']
        
    #     df['predict_sales'] = np.round(self.model.predict([df[['id']], df[['daily_review_count']]]))
    #     # df['predict_sales'] = np.round(df['predict_sales']/3)

    #     df['origin_product_no'] = df['origin_product_no'].apply(int)

    #     return df
    

    def groupby_kpi(self, df, level):
        item_count = df.groupby([f'category{level}_id']).count()[[f'category{level}_name']].reset_index(drop=False)
        item_count.columns = [f'category{level}_id', 'item_count']
        
        df = df.groupby([f'category{level}_id', f'category{level}_name']).apply(self.get_kpi).reset_index(drop=False)
        df['ecommerce'] = 'NAVER'
        if(level <= 2):
            df['is_major'] = 1
            df['gmv_proportion'] = 1
        elif(level > 2):
            df['is_major'] = 0
        else:
            df['is_major'] = None
        df['major_category_name'] = self.ctg
        df['crawling_date'] = pd.to_datetime(pd.to_datetime(self.end_date).strftime("%Y-%m-%d"))
        df['groupby'] = level
        ailee_gmv = self.get_ailee_gmv(level)
        
        
        total_gmv = self.get_ailee_gmv(level+1)['gmv'].sum()
        ailee_gmv[f'category{level}_id'] = ailee_gmv[f'category{level}_id'].apply(str)



        df = df.merge(item_count, on=[f'category{level}_id'], how='left')

        df = df[[f'category{level}_id',
                 f'category{level}_name',
                 'r2',
                 'rmse',
                 'mae',
                 'mape',
                 'mpe',
                 'ecommerce',
                 'is_major',
                 'major_category_name',
                 'crawling_date',
                 'item_count',
                 'groupby'
                 ]]
        
        df = df.merge(ailee_gmv, on=[f'category{level}_id'], how='left')
        df.fillna(0, inplace=True)
        df['gmv_proportion'] = df['gmv']/total_gmv

        # ctg = info[[f'category{level}_id', f'category{level}_name']].drop_duplicates(subset=[f'category{level}_id'])
        # df = df.merge(ctg, on=[f'category{level}_id'], how='left')

        df.rename(columns = {f'category{level}_id': 'category_id',
                             f'category{level}_name': 'category_name'}, inplace=True)
        
        return df
    

    def apply_corr_factor(self, df):
        query = f"""
        SELECT DISTINCT
            category_id AS category4_id,
            mpe
        FROM
            ailee.kpi_predict_6months
        WHERE
            major_category_name = 'pet'
                AND crawling_date = (SELECT 
                    MAX(crawling_date)
                FROM
                    ailee.kpi_predict_6months
                WHERE
                    major_category_name = 'pet')
        """

        conn = self.connect_database(db_name='ailee', user=self.DB_ID, password=self.DB_PASSWORD)

        with conn:
            with conn.cursor(pymysql.cursors.DictCursor) as cursor:
                cursor.execute(query)
                corr = pd.DataFrame(cursor.fetchall())

        corr['category4_id'] = corr['category4_id'].apply(str)
        df = df.merge(corr, on=['category4_id'], how='left')
        df['predict_sales'] = np.round(df['predict_sales'] * df['mpe'])

        return df


    def preprocess(self, info, review):
        selected_columns = ['origin_product_no', 'rank', 'price', 'create_date',
                            'category1_id', 'category2_id', 'category3_id', 'category4_id',
                            'category1_name', 'category2_name', 'category3_name', 'category4_name',
                            'purchaseCnt']
        info = info[selected_columns]
        
        info['origin_product_no'] = info['origin_product_no'].apply(str)
        info.drop_duplicates(subset=['origin_product_no'], keep='last', inplace=True)


        review['origin_product_no'] = review['origin_product_no'].apply(str)
        review['create_date'] = pd.to_datetime(review['create_date'])

        start_date = datetime.now() - timedelta(days=180)

        review = review[review.create_date >= start_date].reset_index(drop=True)

        self.label_encoder = joblib.load('/home/ailee/ailee-v1-ai/DNN/label_encoder.joblib')
        review['origin_product_no'] = review['origin_product_no'].apply(str)

        review = review[review.origin_product_no.isin(self.label_encoder.classes_)]
        review['id'] = self.label_encoder.transform(review['origin_product_no'])

        # dataset = review.merge(info, on=['origin_product_no'], how='left')

        return info, review


    def predict(self, dataset):
        dataset['predict_sales'] = np.round(self.model.predict([dataset['id'], dataset['daily_review_count']]))
        
        return dataset


    def postprocess(self, predict_result, info):
        predict_result = predict_result.groupby(['origin_product_no']).sum()[['predict_sales']].reset_index(drop=False)
        predict_result = predict_result.merge(info, on=['origin_product_no'], how='left')
        predict_result.dropna(subset=['purchaseCnt'], inplace=True)
        predict_result = predict_result[predict_result.purchaseCnt > 0]

        return predict_result
        


    def main(self):
        info = self.get_info()
        target_list = self.get_target_list()
        info = info[info['id'].isin(target_list['id'].unique())]
        review = self.get_review()

        info, review = self.preprocess(info, review)
        predict_result = self.predict(review)
        final_data = self.postprocess(predict_result=predict_result, info=info)

        for level in range(2, 3+1):
            kpi = s.groupby_kpi(df=final_data, level=level)
            s.update_table(kpi)


if __name__ == '__main__':
    # model = tf.keras.models.load_model('/home/ailee/ailee-v1-ai/DNN/training/bestloss_dnn.h5')
    s = ShopperKPI(ctg='pet', model=model, year = 2024, month = 3, day = 7) # 해당 부분을 바꿔가면서 돌리면 됨
    # try:
    #     s.main()
    # except:
    #     s.logger.error(traceback.format_exc())
```
